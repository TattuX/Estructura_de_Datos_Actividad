# Punto 1

1.	Calcular la resistencia total de un circuito compuesto por varias resistencias. Las resistencias tienen diferentes valores, y dependiendo de su valor, pueden sumarse directamente o calcularse en paralelo. El objetivo es ingresar los valores de las resistencias y determinar la resistencia total.
-	Si el valor de la resistencia es menor a 10 o está entre 10 y 100 (inclusive), ese valor se sumará directamente al total.
-	Si el valor de la resistencia es mayor a 100, se calculará su equivalente en paralelo con las otras resistencias que también sean mayores a 100.
La fórmula para resistencias en paralelo es:

image.png


```python


Numero_de_resistencias = int(input("Ingrese cuantas resistencias quiere colocar: "))
Resistencias =  []

for i in range(Numero_de_resistencias):
    resistencia = float(input(f"Ingresar el valor de la resistencia {i+1} :"))
    Resistencias.append(resistencia)

Total_resistencias = 0

Resistencias_en_serie = []

Resistencias_en_paralelo = []

for resistencia in Resistencias:

  if resistencia < 10 or (resistencia > 10 and resistencia < 100):
    Resistencias_en_serie.append(resistencia)
  else:
    Resistencias_en_paralelo.append(resistencia)

Total_resistencias = sum(Resistencias_en_serie)

if Resistencias_en_paralelo:
   Resistencia_total_del_paralelo = 1 / sum(1/ r for r in Resistencias_en_paralelo)
else:
  Resistencia_total_del_paralelo = 0

resistencia_total = Total_resistencias + Resistencia_total_del_paralelo

print(f"La resistencia total es: {resistencia_total:.2f} ohmnios")

```
